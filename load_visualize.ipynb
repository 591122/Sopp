```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a Pandas DataFrame
file_path = 'mushrooms.csv'  # Replace 'mushrooms.csv' with the actual file path
df = pd.read_csv(file_path)

# Visualize the data for all attributes, splitting by 'class' (poisonous vs. not poisonous)
sns.set(style="whitegrid")  # Set the style for Seaborn

# List of attributes to plot (excluding the 'class' column)
attributes = df.columns.drop('class')

# Create subplots for each attribute
fig, axes = plt.subplots(nrows=len(attributes), ncols=1, figsize=(10, 6 * len(attributes)))

for i, attribute in enumerate(attributes):
    ax = axes[i]
    sns.countplot(data=df, x=attribute, hue='class', ax=ax)
    ax.set_title(f"Distribution of {attribute}")
    ax.set_xlabel(attribute)
    ax.set_ylabel("Count")
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
